1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou:

```python
x = sp.symbols('x')
f = x**3
resultado = sp.integrate(f, (x, 1, 2))

print(f'O valor da integral é: {resultado}')
```

```python
from sympy import Symbol, plot

x = Symbol('x')
expr = x**3

plot(expr, (x, 1, 2));
```

```python
import numpy as np
import matplotlib.pyplot as plt

# definindo a função
f = lambda x: x**3

# intervalo
a = 1; b = 2;

# numero de retangulos
N = 10

n = 10 # Use n*N+1 points to plot the function smoothly

# discretizando variáveis
x = np.linspace(a,b,N+1)
y = f(x)
X = np.linspace(a,b,n*N+1)
Y = f(X)

# iniciando plot
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(X,Y,'b')
x_left = x[:-1]
y_left = y[:-1]
plt.plot(x_left,y_left,'b.',markersize=10)
plt.bar(x_left,y_left,width=(b-a)/N,alpha=0.2,align='edge',edgecolor='b')
plt.title('Soma de Riemann com base na esquerda'.format(N))

plt.show()
```

2- Calcule essa área de forma analítica:

```python
dx = (b-a)/N
x_left = np.linspace(a,b-dx,N)

print("Partition with",N,"subintervals.")
left_riemann_sum = np.sum(f(x_left) * dx)
print("Left Riemann Sum:",left_riemann_sum)
```
